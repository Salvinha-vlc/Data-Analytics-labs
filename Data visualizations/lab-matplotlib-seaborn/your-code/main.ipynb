```python
fitbit2.Months_encoded.describe()
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
median_by_month = fitbit2.pivot_table(index=["Months_encoded"], values=["Steps"],aggfunc="median").reset_index()

median_by_week_weekday = fitbit2.pivot_table(index=["Work_or_Weekend"], values=["Steps"],aggfunc="median").reset_index()
```

```python
median_by_month
```

```python
median_by_week_weekday
```

```python
fig, [ax1, ax2] = plt.subplots(2,1,figsize=(8,10))
ax1.bar(median_by_month.Months_encoded,median_by_month.Steps)
ax1.set_title('Median steps by month')   
ax1.set_xlabel('Months')     
ax1.set_ylabel('Steps')  

labels=["Weekend","Weekday"]

ax2.bar(median_by_week_weekday.Work_or_Weekend,median_by_week_weekday.Steps, tick_label=labels, color="darksalmon", width=0.5)
ax2.set_title('Median by weekday')
ax2.set_xlabel("day type")
ax2.set_ylabel('Steps')

plt.savefig('../figures/median_steps_by_month_and_weekday.png', dpi=300)
plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
sns.heatmap(numerical_titanic.corr(min_periods=12), annot=True)
plt.show()
```

```python
# another more elaborated way
```

```python
# your code here
corr = numerical_titanic.corr()
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))
# Generate a custom diverging colormap
cmap = sns.diverging_palette(203, 10, as_cmap=True)
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
           square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

#### What are the most correlated features?
